```python
# importing all the requuired packages

import pandas as pd
import numpy as np
```

```python
# grabbing the dataset of kaggle
los_data = pd.read_csv("./Data/kaggle_los.csv")
los_data.head()
```

# Cleaning the data

```python
# cgetting a sample data

los_data.sample(5)
```

```python
# checking for missing values

missing_values_count = los_data.isnull().sum()
missing_values_count
```

```python
# how many total missing values do we have?
total_cells = np.product(los_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
# 6% data is empty
```

```python
# look at the # of missing points in the first ten columns
missing_values_count[0:10]
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = los_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# checking for missing values

missing_values_count = columns_with_na_dropped.isnull().sum()
missing_values_count
```

```python
# how many total missing values do we have?
total_cells = np.product(columns_with_na_dropped.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```
